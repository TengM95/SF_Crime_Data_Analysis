## San Francisco crime data analysis and modeling

### Abstract
In this project, we processed San Francisco crime data from 2003 to 2018 from San Francisco Police Department https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-Historical-2003/tmnf-yvry.  

Did data cleaning, data visualization by Spark Dataframe, Spark SQL for big data OLAP and analyzed the distribution of SF crime over category, district, hour, day and month separately. Then we used k-means based on Spark ML to explore and visualize the spatial distribution of crime over the category and year.

#### Count the number of crimes each Sunday at SF downtown.  
The center of downtown is (37.778548, -122.418228) and the farest point of downtown is (37.782771, -122.414076).

```
#Spark dataframe
# isDowntown_udf_bool
def isDowntown(Y, X):
  return (float(Y) - 37.778548) **2 + (float(X) + 122.418228) **2 <= (37.778548 - 37.782771) **2 + (-122.418228 + 122.414076) ** 2
isDowntown_udf_bool = udf(lambda Y,X: isDowntown(Y, X), BooleanType())


q3_res = df_opt1.where(isDowntown_udf_bool('Y', 'X') & (df_opt1.DayOfWeek == 'Sunday'))\
                .groupBy('Date')\
                .count()\
                .orderBy(to_date('Date','MM/dd/yy'))
display(q3_res)
```

```
#Spark sql 
spark.udf.register("isDowntown_udf_bool", isDowntown, BooleanType())
q3_res = spark.sql("SELECT Date, count(*) as numOfCrime  FROM sf_crime where isDowntown_udf_bool(Y, X) AND DayOfWeek = 'Sunday' GROUP BY Date ORDER BY to_date(Date,'MM/dd/yy')")

display(q3_res)
```

From above bar chart, we can see the number of crime events of last sunday of June is the largest over the year.   
The day is San Francisco Pride day which may be the reason for this observation.

#### Analysis the number of crime in each month of 2014, 2015, 2016, 2017.

```
#Spark dataframe
df_opt1_new = df_opt1.withColumn('Date', to_date('Date','MM/dd/yy'))\
                    .select('*', date_format('Date', 'MM').alias('Month'), date_format('Date', 'yyyy').alias('Year'))
df_opt1_new.createOrReplaceTempView('sf_crime_date')

q4_res = df_opt1_new.groupBy('Year', 'Month').count().where(col('Year').isin(['2015', '2016', '2017', '2014'])).orderBy('Year', 'Month')
display(q4_res)
```

```
#Spark sql 
q4_res = spark.sql("SELECT date_format(Date,'yyyy') as Year, date_format(Date,'MM') as Month, COUNT(IncidntNum) AS numOfCrime  FROM sf_crime_date GROUP BY 2,1  HAVING Year in ('2015', '2016', '2017', '2014') ORDER BY 1,2")

display(q4_res)
```

We can see the amount of crime of each month in 2018 is less than before. And the amount of crime in March and October is usually higher than other months. The police should put more police strength on March.
